```python
from bokeh.io import curdoc, show, output_notebook
from bokeh.layouts import column, row, widgetbox, gridplot
from bokeh.models.widgets import TextInput, Button, Paragraph, Slider, Div

def modify_doc(doc):
    output = Paragraph()
    text_query = TextInput(value = "random", width=220, height=10, title = "Enter article query")
    button_query = Button(label = 'Search articles', width=220, height=10, button_type='success')
    empty_div = Div(text="", width=70, height=1)
    query_grid = row(text_query, column(empty_div, button_query))
    
    text_user = TextInput(placeholder='Enter your name', value = 'saket', width = 340)
    username = Div(text="Username:", width=70, height=10)
    
    rec_button = Button(label = 'Get recommendations', width=110, height=5, button_type='warning')
    reset_button = Button(label = 'Reset Preferences', width = 150, height = 5, button_type = 'warning')
    user_grid1 = row(column(empty_div, username), text_user)
    user_grid2 = row(row(Div(text="", width = 70), rec_button), row(Div(text="", width = 5), reset_button))
    user_grid = column(user_grid1, user_grid2)
    
    grid = row(query_grid, Div(text="", width = 60), user_grid)
    doc.add_root(grid)

    def update():
        output.text = "Hello, " + text_query.value

    button_query.on_click(update)

output_notebook()
show(modify_doc)
```

```python
import yaml

from bokeh.layouts import column
from bokeh.models import ColumnDataSource, Slider
from bokeh.plotting import figure
from bokeh.themes import Theme
from bokeh.io import show, output_notebook

from bokeh.sampledata.sea_surface_temperature import sea_surface_temperature

output_notebook()

def modify_doc(doc):
    df = sea_surface_temperature.copy()
    source = ColumnDataSource(data=df)

    plot = figure(x_axis_type='datetime', y_range=(0, 25),
                  y_axis_label='Temperature (Celsius)',
                  title="Sea Surface Temperature at 43.18, -70.43")
    plot.line('time', 'temperature', source=source)

    def callback(attr, old, new):
        if new == 0:
            data = df
        else:
            data = df.rolling('{0}D'.format(new)).mean()
        source.data = ColumnDataSource(data=data).data

    slider = Slider(start=0, end=30, value=0, step=1, title="Smoothing by N Days")
    slider.on_change('value', callback)

    doc.add_root(column(slider, plot))

    doc.theme = Theme(json=yaml.load("""
        attrs:
            Figure:
                background_fill_color: "#DDDDDD"
                outline_line_color: white
                toolbar_location: above
                height: 500
                width: 800
            Grid:
                grid_line_dash: [6, 4]
                grid_line_color: white
    """))
    
    
    
show(modify_doc)
```
